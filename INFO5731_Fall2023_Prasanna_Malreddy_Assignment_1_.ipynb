# Question 1

(20 points). Write a Python program to generate 12 random numbers between 1 and 100. Sort the list of numbers in ascending order and print it out. Sort the list of numbers in descending order and print it out. Calculate the average for the group and print it out.

```python
import random
Num=[]
for i in range(12):
    Num.append(random.randint(1,100))
print(Num)
Num.sort()
print("Numbers in ascending Order:",Num)
Num.sort(reverse=True)
print("Numbers in descending Order:",Num)
avg=sum(Num)/len(Num)
print("Average of the list of numbers:",avg)

```

# Question 2
(25 points).

Write a program that will do the following string formatting tasks:

(1) Ask the user to enter two sentences.

(2) Make each sentence into a list. Each element in the list should correspond to a word in the sentence. For example, if the sentence is "Alas! Am I alive? This is my first python program", the program should print out ['Alas, '!', 'Am', 'I', 'alive, '?', 'This', 'is', 'my', 'first', 'python', 'program'].

(3) Compare the two sentences. Print out a list of words that occur in both sentences.

(4) Remove any punctuation from the sentence and print it back out. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”)

(5) Count the number of vowels in the sentences. Print out each vowel and the number of times it appears in the sentences, such as: a:2, e:1, i:0, etc

```python
import string
import re

str1=input("Enter first sentence: ")
str2=input("Enter second sentence: ")
# Converting the sentence into a list of words
list1 = [word for word in re.findall(r'\b\w+\b|[^\w\s]', str1)]
list2 = [word for word in re.findall(r'\b\w+\b|[^\w\s]', str2)]
print("First sentence into List: ",list1)
print("Second sentence into List: ",list2)

word_list1 = [word.lower() for word in re.findall(r'\b\w+\b', str1)]
word_list2 = [word.lower() for word in re.findall(r'\b\w+\b', str2)]
common_words = [word for word in word_list1 if word in word_list2]
print("Words that occur in both sentences:", common_words)


print("First sentence without punctuation:"," ".join(word_list1))
print("Second sentence without punctuation:"," ".join(word_list2))


def count_vowels(sentence):
    vowels = "aeiou"
    vowel_count = {v:0 for v in vowels}
    for char in sentence:
        if char in vowels:
            vowel_count[char] += 1

    for vowel, count in vowel_count.items():
        print(f"{vowel}: {count}")
print("Vowels in First Sentence: ")
count_vowels(str1.lower())
print("Vowels in Second Sentence: ")
count_vowels(str2.lower())
```

(15 points). The formula that describes the volume V of a sphere with radius r is the following:

$ V=\frac{4}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of 1–10. Output the result in the following format:

	r	V
	1	…
	2	…
	3	…


```python

import math

print("r\tV")
for r in range(1, 11):
    V = (4/3)*math.pi*(r**3)
    print("{}\t{:.2f}".format(r,V))
```

```python
import pandas as pd
dataframe = pd.read_csv('Assignment1_denton_housing.csv')
dataframe
```

```python
# 1) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order
occupied=dataframe.loc[dataframe['title_field'] == 'Occupied Housing Units']
occupied
```

```python
#Calculating the difference in number of occupied housing units for each consecutive years and printing them in ascending order
```

```python
occupied_vals=occupied['value'].values
diffs_a=occupied_vals[:-1]-occupied_vals[1:]
years=["2014-2013","2013-2012","2012-2011","2011-2010","2010-2009","2009-2008"]
dataframe2=pd.DataFrame()
dataframe2['years']=years
dataframe2['Difference']=abs(diffs_a)
print(dataframe2)
print("Sorting the difference from year to year: ")
sorted_df = dataframe2.sort_values(by='Difference')
print(sorted_df)
```

```python
#(2)For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format: Year Vacant Housing Units Occupied Housing Units
```

```python
print(occupied)
print(occupied_vals)
vacant=dataframe.loc[dataframe['title_field']=='Vacant Housing Units']
vacant_vals=vacant['value'].values
print(vacant)
print(vacant_vals)
total=dataframe.loc[dataframe['title_field']=='Total Housing Units']
tot_vals=total['value'].values
print(total)
print(tot_vals)
perc_of_occupied=[]
perc_of_vacant=[]
for i in range(7):
    perc_of_occupied.append(round(((occupied_vals[i]/tot_vals[i])*100),2))

for i in range(7):
    perc_of_vacant.append(round(((vacant_vals[i]/tot_vals[i])*100),2))

year=occupied['year'].values
print("Percentage of Vacant and Occupied Housing units for all years: ")
print("Year\tVacant Housing Units\tOccupied Housing Units")
for i in range(7):
    print("{}\t\t{:.2f}\t\t{:.2f}".format(year[i],perc_of_vacant[i],perc_of_occupied[i]))
```
